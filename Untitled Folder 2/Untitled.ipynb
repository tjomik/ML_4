```python
data_train = pd.read_csv('train.csv', delimiter=',', index_col='id')
data_test = pd.read_csv('test.csv', delimiter=',', index_col='id')
```

```python
data_train_count_y0 = data_train[data_train['Occupancy']==0].shape[0]
data_train_count_y1 = data_train[data_train['Occupancy']==1].shape[0]
data_train_count_y0
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, validation = train_test_split(data_train, test_size=0.4)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x_columns=['CO2', 'Light','Temperature', 'Humidity']
y_column='Occupancy'
```

```python
a=LogisticRegression().fit(train[x_columns], train[y_column])
y_train = a.predict(train[x_columns])
y_acc = accuracy_score(y_train, train[y_column])
y_acc
```

```python
y_val = a.predict(validation[x_columns])
y_acc_val = accuracy_score(y_val, validation[y_column])
y_acc_val
```

```python
y_pred = a.predict(data_test[x_columns])
```

```python
data_test['value']=y_pred

data_test.to_csv('result.csv')
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(train[x_columns], train[y_column])
y_test_2 = clf.predict(validation[x_columns])
y_test_acc= accuracy_score(y_test_2,validation[y_column])
print(y_test_acc)
occupancy_predicted = clf.predict(data_test[x_columns])
```

```python
data_test['value']=occupancy_predicted
data_test.to_csv('result.csv')
```
